```python
#기상데이터 수정 code
import pandas as pd
import os

# 1. GitHub Raw URL
url = "https://raw.githubusercontent.com/JJINwj/Time-series-demand-forecasting-team2/main/data/raw/01.bike/%EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C%20%EA%B8%B0%EC%83%81%EB%8D%B0%EC%9D%B4%ED%84%B0/OBS_ASOS_TIM_201701-12.csv"

# 2. 데이터 불러오기
df = pd.read_csv(url, encoding="cp949")

# 3. datetime 변환
df['일시'] = pd.to_datetime(df['일시'], format='%Y-%m-%d %H:%M')

# 4. 날짜/시간 분리
df['날짜'] = df['일시'].dt.strftime('%Y-%m-%d')
df['시간'] = df['일시'].dt.strftime('%H')

# 5. 필요한 변수만 선택
cols = [
    '날짜', '시간',
    '기온(°C)',
    '강수량(mm)',
    '풍속(m/s)',
    '습도(%)',
    '시정(10m)'
]
df_out = df[cols]

# 6. 저장 경로 설정
save_path = "C:/Users/user/Documents/code/Time-series-demand-forecasting-team2/data/interim/01.bike/modi_weather_2017.csv"

# 7. 저장
df_out.to_csv(save_path, index=False, encoding="cp949")

print("파일 저장 완료:", save_path)
```
