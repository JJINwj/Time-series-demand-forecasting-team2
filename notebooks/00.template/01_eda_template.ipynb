```python
#1. 라이브러리 & 설정
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (10, 4)
sns.set_style("whitegrid")
```

```python
#2. 데이터 불러오기(예시)
# 예: 따릉이 수요, 도시철도 수요, 비트코인 가격 등
ata_path = "data/raw/01.bike/bike_raw.csv"  # 각자 파일 경로로 수정
df = pd.read_csv(data_path)

# datetime 컬럼 파싱 (컬럼 이름도 각자 수정)
df["datetime"] = pd.to_datetime(df["datetime"])
df = df.sort_values("datetime").reset_index(drop=True)

df.head() #데이터 확인
```

```python
# 기본 정보
print(df.shape)
print(df.info())
df.describe(include="all")
```

```python
# 3. 결측치 확인
df.isnull().sum().sort_values(ascending=False)
```

```python
# 숫자형 컬럼 분포 확인 (히스토그램)
num_cols = df.select_dtypes(include=["float64", "int64"]).columns
df[num_cols].hist(bins=30, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
# 4. 시계열 패턴 – 전체 추이
target_col = "demand"  # 예: 따릉이 이용량, 철도 승차인원 등
plt.plot(df["datetime"], df[target_col])
plt.title(f"{target_col} – time series")
plt.xlabel("datetime")
plt.ylabel(target_col)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# 요일/시간대 등 파생변수 잠깐 만들어서 패턴만 보기
df["dayofweek"] = df["datetime"].dt.dayofweek
df["hour"] = df["datetime"].dt.hour

# 요일별 평균
df.groupby("dayofweek")[target_col].mean().plot(kind="bar")
plt.title(f"Average {target_col} by dayofweek")
plt.show()

# 시간대별 평균
df.groupby("hour")[target_col].mean().plot(kind="bar")
plt.title(f"Average {target_col} by hour")
plt.show()
```
